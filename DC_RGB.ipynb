```python
from Model import Basic_RGB
from Model import Evaluation
from Data import load_data
```

```python
path_training_data = "Data/WM18_train.pkl"
path_testing_data = "Data/WM18_test.pkl"
```

```python
Data_training = load_data.load_obj(path_training_data)
Data_testing = load_data.load_obj(path_testing_data)
Rmu, Wordvector, Tmu = load_data.training_data(Data_training)
E_Rmu, E_Wordvector, E_Tmu = load_data.testing_data(Data_testing)
```

```python
dc_model =  Basic_RGB.DC_RGB()
```

```python
train_tuples = [i[:3] for i in Data_training]
test_tuples = [i[:3] for i in Data_testing]
```

```python
result = {}
```

```python
n = 10
for i in range(n):
    history = dc_model.fit([Rmu, Wordvector], Tmu, epochs=50, batch_size=16)

    Pred = dc_model.predict([E_Rmu, E_Wordvector])
    
    result[i] = Evaluation.evaluation(train_tuples, test_tuples, Pred, E_Rmu, E_Tmu)
```

```python
keys = ["Seen_pairings",
        "Unseen_Pairings",
        "Unseen_Reference",
        "Unseen_Modifier",
        "Fully_Unseen",
        "Overall"
        ]
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))

x = [i*50 for i in range(n)]

plt.subplot(2, 1, 1)
for k in keys:
    plt.plot(x, [result[i][0][k] for i in range(n)], '.-', label = k)
plt.title('Training history')
plt.ylabel('Cosine similarity')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.subplot(2, 1, 2)
for k in keys:
    plt.plot(x, [result[i][1][k] for i in range(n)], '.-', label = k)
plt.xlabel('epoches')
plt.ylabel('Delta-E')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

```python
Pred = dc_model.predict([E_Rmu, E_Wordvector])
Cosin, Delta = Evaluation.evaluation(train_tuples, test_tuples,
                                     Pred, E_Rmu, E_Tmu)
```
