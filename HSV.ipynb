```python
from Model import HSV
from Model import Evaluation
from Data import load_data
import numpy as np
```

```python
keys = ["Seen_pairings",
        "Unseen_Pairings",
        "Unseen_Reference",
        "Unseen_Modifier",
        "Fully_Unseen",
        "Overall"
        ]
```

```python
path_training_data = "Data/WM18_train.pkl"
path_testing_data = "Data/WM18_test.pkl"
```

```python
Data_training = load_data.load_obj(path_training_data)
Data_testing = load_data.load_obj(path_testing_data)
Rmu, Wordvector, Tmu = load_data.training_data(Data_training)
E_Rmu, E_Wordvector, E_Tmu = load_data.testing_data(Data_testing)
```

```python
from matplotlib.colors import hsv_to_rgb
from  matplotlib.colors import rgb_to_hsv
```

```python
Rmu = np.array([rgb_to_hsv(i) for i in Rmu])
Tmu = np.array([rgb_to_hsv(i) for i in Tmu])
```

```python
Tmu_H = np.array([i[0] for i in Tmu])
Tmu_SV = np.array([i[1:] for i in Tmu])
```

```python
E_Rmu = np.array([rgb_to_hsv(i) for i in E_Rmu])
```

```python
train_tuples = [i[:3] for i in Data_training]
test_tuples = [i[:3] for i in Data_testing]
```

```python
model =  HSV.HSV()
```

```python
result = {}

n = 40

for i in range(n):
    history = model.fit([Rmu, Wordvector], 
                       [Tmu_H, Tmu_SV],
                       epochs=50,
                       batch_size=32)
    Pred = model.predict([E_Rmu, E_Wordvector])
    Pred = [ [Pred[0][i][0], Pred[1][i][0], Pred[1][i][1]] for i in range(Pred[0].shape[0])]
    Pred = [hsv_to_rgb(i) for i in Pred]
    result[i] = Evaluation.evaluation(train_tuples, test_tuples,
                                     Pred, E_Rmu, E_Tmu)
```
